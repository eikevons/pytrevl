# Sort data for line-chart

```python
from pytrevl import LineChart, CubeQuery
```

```python
_q = CubeQuery(
    'test_hrm',
    ['answer_count'],
    ['question_key'],
)

line_asc = LineChart(_q)
print(line_asc.as_yaml())
line_asc.show()
```

```python
line_desc = LineChart(_q, order_direction='desc')
print(line_desc.as_yaml())
line_desc.show()
```
